```
import numpy as np
import matplotlib.pyplot as plt

lang = ["Python", "C++", "C", "R"]
number = [50, 80, 70, 60]
y = [20, 80, 70, 45]
p = np.arange(len(lang))
width = 0.3  # Increased width for better separation
p1 = p + width

plt.bar(p, number, width=width, align="center", alpha=0.7, label="Language 1")
plt.bar(p1, y, width=width, align="center", label="Language 2")
plt.xticks(p + width / 2, lang, rotation=20)

#plt.xlabel("Language")
#plt.ylabel("Number of Users")
plt.legend()
plt.title("Popularity", fontsize=15)
plt.show()
```
