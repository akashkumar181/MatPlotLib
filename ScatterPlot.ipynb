```python
import matplotlib.pyplot as plt
```

Analyzing the Gdp of India through scatter plot in different year

```python
year=[2000,2001,2015,2003,2017,2025,2010]
gdpindia=[468.4,485.44,2103.59,709.15,2651.47,3549.92,1675.62]
gdpamerica=[10250000,10580000,15050000,183000000,19610000,27360000,87800000]
size=[100,100,100,100,100,100,100]
plt.scatter(year,gdpindia,s=size,alpha=0.7,color='y')
plt.scatter(year,gdpamerica,s=size,alpha=0.7,color='g')
#plt.bar(year,gdpindia,alpha=0.7,color='y')
plt.title("GDP VARIATION OF INDAIA IN DIFFERENT YEAR",color='g')
plt.xlabel("Year",color='r')
plt.ylabel("GDP in Billion USD",color='r')
plt.show()
```
