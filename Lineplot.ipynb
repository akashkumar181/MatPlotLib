#Line plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


```

#Basic Graph

```python
x=[1,2,3,4,5]
y=[0,2,4,6,10]
#plt.plot(x,y,label="2x",color='r',linewidth=2,marker="*",markersize=10,markeredgecolor='b',linestyle="--")
plt.title("Line plot",fontdict={'fontsize':20})
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.legend()
plt.xticks([0,1,2,3,4])#by this we remove decimal value from x axis
plt.yticks([0,2,4,6,8,10])
plt.show()
```

```python
x=[0,1,2,3,4,5]
y=[0,2,4,6,8,10]
plt.title("Line plot",fontdict={'fontsize':20})
plt.xlabel("x-axis")
plt.ylabel("y-label")
x2=np.arange(0,4.5,0.5)
#plt.plot(x2,x2**2,"r*--",label="x2^2")
plt.plot(x2[:6], x2[:6]**2,"r--",label="x2^2")
plt.plot(x2[5:],x2[5:]**2)

plt.plot(x,y,"g*--",label="2x")
plt.legend()
plt.xticks([0,1,2,3,4])#by this we remove decimal value from x axis
plt.yticks([0,2,4,6,8])
plt.figure(figsize=(10,20),dpi=200)
plt.savefig("mygraph.png",dpi=300)
plt.show()
```

#Bar chart

```python
labels=['a','b','c']
values=[1,4,6]
bars=plt.bar(labels,values,label="BAR CHART",color='pink')
bars[0].set_hatch('/')
bars[1].set_hatch("//")
bars[2].set_hatch("*")
plt.legend()
plt.figure(figsize=(10,30),dpi=300)
plt.show()
```
