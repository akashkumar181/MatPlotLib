```
import numpy as np
import matplotlib.pyplot as plt
Tata=[667000,650000,749000]#inr
Amazon=[107000,135900,280500] #us dollar
finacial_year=[2015,2016,2019]
p=np.arange(len(finacial_year))
width=0.4
p1=p+width
#plt.xlabel("Fincial Year")

#plt.ylabel("Revenue")

plt.bar(p,Tata,width=width,label="Tata")
plt.bar(p1,Amazon,width=width,label="Amazon")
plt.xticks(p+width/2,finacial_year,rotation=20)
plt.legend()
plt.title("Finacial Status of two company",color='g')
plt.show()
```
