```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
year=[2020,2021,2022,2023,2024]
tempIndia=[1.02,1.09,1.12,1.15,1.18]
tempAmerica=[1.00,1.05,1.08,1.10,1.12]
plt.plot(year,tempIndia,"--.",label='Tem Rise in India')
plt.plot(year,tempAmerica,"--.",label="Tem Rise in America")
plt.title("Temperatue Rise",color='g')
plt.xlabel("Year")
plt.xticks([2020,2021,2022,2023,2024])
plt.yticks([1.00,1.02,1.04,1.06,1.08,1.10,1.12,1.14])
plt.ylabel("Temperature Rise")
plt.grid()
plt.tight_layout()
plt.legend()
plt.show()
```
